```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
csv_file = ["TheHalving_random.csv",
            "TheHalving_human.csv",
            "Nim_random.csv",
            "Nim_human.csv",
            "TicTacToe_random.csv",
            "TicTacToe_human.csv",
            "ConnectFour_random.csv",
            "ConnectFour_human.csv"
           ]

plt.rcParams.update({'font.size': 8})
```

```python
#win_rate_comparison
random_df = pd.read_csv(csv_file[0])

human_df = pd.read_csv(csv_file[1])

halving_random_minimax_rate = random_df["minimax_bot_wins"].sum() / len(random_df["minimax_bot_wins"])
halving_random_rate = (len(random_df["minimax_bot_wins"]) - random_df["minimax_bot_wins"].sum()) / len(random_df["minimax_bot_wins"])

halving_human_minimax_rate = human_df["minimax_bot_wins"].sum() / len(human_df["minimax_bot_wins"])

halving_human_rate = (len(human_df["minimax_bot_wins"]) - human_df["minimax_bot_wins"].sum()) / len(human_df["minimax_bot_wins"])


labels = ["Minimax Bot Wins\n(Against Human)", "Human Win Rate","Minimax Bot Wins\n(Against Random)", "Random Agent Win Rate"]
sizes = [halving_human_minimax_rate, halving_human_rate, halving_random_minimax_rate,halving_random_rate]
plt.bar(labels,sizes, color=["blue", "orange","green","red"], width= 0.6)
plt.title("Minimax Bot Win Rates (Halving, 50 games with Human, 10000 games with Random Agent)")
plt.savefig('halving_win_rates.png')
plt.show()

```

```python
#distribution_of_moves


halving_rand_move_labels = random_df["turns"].unique()
rand_move_cnt = []
for label in halving_rand_move_labels:
    rand_move_cnt.append(len(random_df[random_df["turns"] == label]))

plt.bar(halving_rand_move_labels, rand_move_cnt, color="blue", width= 0.6)
plt.title("Distribution of Moves in Random Agent Games (Halving, 10000 games)")
plt.xticks(np.arange(0, 21, step=1))
plt.savefig('halving_random_distribution.png')
plt.show()


halving_human_move_labels = human_df["turns"].unique()
halving_human_move_cnt = []
for label in halving_human_move_labels:
   halving_human_move_cnt.append(len(human_df[human_df["turns"] == label]))

plt.bar(halving_human_move_labels, halving_human_move_cnt, color="blue", width= 0.6)
plt.title("Distribution of Moves in Human Games (Halving, 50 games)")
plt.savefig('halving_human_distribution.png')
plt.xticks(np.arange(0, 21, step=1))
plt.show()

```

```python
four_random_df = pd.read_csv(csv_file[6])

four_minimax_random_win_rate = four_random_df["bot_wins"].sum()/len(four_random_df["bot_wins"])
four_tie_rate = len(four_random_df[four_random_df["winner"] == "Draw"])/len(four_random_df["winner"])
four_random_win_rate = 1- four_minimax_random_win_rate - four_tie_rate

four_human_df = pd.read_csv(csv_file[7])

four_minimax_human_win_rate = four_human_df["bot_wins"].sum()/len(four_human_df["bot_wins"])
four_human_tie_rate = len(four_human_df[four_human_df["winner"] == "Draw"])/len(four_human_df["winner"])
four_human_win_rate = 1- four_minimax_human_win_rate - four_human_tie_rate


four_labels = ["Minimax Bot Wins\n(Against Human)", "Human\nWin Rate","Human\n Tie Rate","Minimax Bot Wins\n(Against Random)", "Random\nWin Rate","Random\nTie Rate"]
four_sizes = [four_minimax_human_win_rate, four_human_win_rate, four_human_tie_rate, four_minimax_random_win_rate, four_random_win_rate, four_tie_rate]

plt.bar(four_labels,four_sizes, color=["blue", "orange","green","red"], width= 0.3)
plt.title("Minimax Bot Win Rates\n(Connect Four, 50 games with Human, 10000 games with Random Agent)")
plt.savefig('four_win_rates.png')
plt.show()


```
